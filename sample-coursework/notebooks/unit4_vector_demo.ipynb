# Unit 4 vector evidence

This notebook records a small vector similarity check for the A.L.I. sample coursework folder.

```python
import math

lost_item = [0.9, 0.2, 0.1, 0.4]
found_item = [0.8, 0.3, 0.2, 0.5]
dot = sum(a * b for a, b in zip(lost_item, found_item))
mag_a = math.sqrt(sum(a * a for a in lost_item))
mag_b = math.sqrt(sum(b * b for b in found_item))
print('cosine similarity:', round(dot / (mag_a * mag_b), 2))
```
